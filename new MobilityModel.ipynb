As can be seen, the dog's movement was simulated through a sine wave. The x axis in this sine function is the firefighter who moves in a straight line. Now, about the speed of the dogs: of course it should be randomized, but besides that, to make it so that the dogs follow the firefighters (i.e. they move through the y axis more or less together) it is necessary to calculate the length of the sine wave. Introduced deviations from the sine wave. Every time a dog updates it's position there is a probability of it straying away from the path. Looks a little bit rough because there are no intermidiate points between the deviation and the dog's current position as can be seen in the fact that not all the dogs end up in the same y position despite having the same speed.

To control the dogs' speed and position when gone astray: say the dog's current position is (x,y) and the goal position is (xF, yF), then a versor should be calculated, like this
$$ \begin{align}
xDir = xF - x\\
yDir = yF - y\\
xVersor = xDir / \sqrt{ xDir^2 + yDir^2} \\
yVersor = yDir / \sqrt{ xDir^2 + yDir^2}\\
\end{align}$$
Then, to reach the final position in each timeStep: $$ nextX = previousX + timeStep * nodeSpeed * xVersor$$ and the same goes for nextY

In order to make the speed of the dogs along the path behave nonuniformly, these are drawn from a cuasi normal distribution, with mean in DOG_SPEED and a standar deviation given by the parameter STD_DEVIATION.
It should be noted that since the probability of drawing a really high speed is not null in a normal distribution, all the numbers drawn lie between (DOG_SPEED - STD_DEVIATION, DOG_SPEED + STD_DEVIATION). Hence, cuasi normal distrubution.

### Obstacles

An obstacle is defined as a line segment given by two points (x1,y1) and (x2,y2). Several line segments can form a polygon or not, that's left to the user. 
An obstacle definition is given in a separated file (-o flag) as a list of floats delimited by commas. Like this:


That would define two line segments in (1.0;2.3), (3.1,4) and (5,6);(7,8).\n
Any node would have to notice an obstacle as soon as it enters its line of sight, then ask itself if given its current movement, a collision might happen. If yes, calculate the shortest path to avoid collision.
While avoiding an obstacle, both ff and dogs would leave their charachteristic movement, at least at first.

After _cf47038600_, when an obstacle gets in the line of sight of a firefighter it will calculate the closest vertex of the obstacle and the it will head into a direction given by (closestVertexX +- 5; closestVertexY +-5) as long as that position lies inside the field. 

If there are no obstacles in the new calc position, the ff will head in that direction. It will then try to return to its original position and if not possible (because there is an obstacle in the way) it will continue until the way is clear. The dog interrupts its movement and tags along the ff.
