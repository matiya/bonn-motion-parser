As can be seen, the dog's movement was simulated through a sine wave. The x axis in this sine function is the firefighter who moves in a straight line. Now, about the speed of the dogs: of course it should be randomized, but besides that, to make it so that the dogs follow the firefighters (i.e. they move through the y axis more or less together) it is necessary to calculate the length of the sine wave. This is done with the following integral:
$$DELTA\_X ~ *~ \int_a^{b} \sqrt{1+\cos^2 x}\,\mathrm dx$$ I don't know how to do this in c++, so unfortunately we'll have to go with some SAC and compute the result, like:

Introduced deviations from the sine wave. Every time a dog updates it's position there is a probability of it straying away from the path. Looks a little bit rough because there are no intermidiate points between the deviation and the dog's current position as can be seen in the fact that not all the dogs end up in the same y position despite having the same speed.

To control the dogs' speed and position when gone astray: say the dog's current position is (x,y) and the goal position is (xF, yF), then a versor should be calculated, like this
$$ \begin{align}
xDir = xF - x\\
yDir = yF - y\\
xVersor = xDir / \sqrt{ xDir^2 + yDir^2} \\
yVersor = yDir / \sqrt{ xDir^2 + yDir^2}\\
\end{align}$$
Then, to reach the final position in each timeStep: $$ nextX = previousX + timeStep * nodeSpeed * xVersor$$ and the same goes for nextY

```python
import sys
import os
import getopt
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.lines import Line2D
from matplotlib import animation
from IPython.display import Image

#global options
pylab.rcParams['figure.figsize'] = 16, 12
fig = plt.figure()
ax = plt.axes(xlim=(0, 400), ylim=(0, 400))
line, = ax.plot([], [], lw=2)

def oneNode(opt, fileName):
  """plot one node"""
  ax = fig.add_subplot(1,1,1)
  a = np.loadtxt(str(fileName[1]))
  x = a[:,0]
  y = a[:,1]
  ax.plot(x,y,'--')
  if opt == 1:
    plt.show()
    
def allNodesIn(cwd):
  """plots all the .visplotX files found in the given dir
  """
  legendList = []
  print("found the following:")
  for argument in os.listdir(cwd):
    if 'visplot' in argument:
      oneNode(2,[1,str(cwd)+"/"+argument])
      print(str(cwd)+"/"+argument)
      legendList.append(str(argument))
  ax = fig.add_subplot(1,1,1)
  ax.legend(legendList)
  plt.draw()
  plt.show()

def allNodes(fileName):
  """plots all the nodes, the input is the decompressed file produced by bm"""
  fig1 = plt.figure()
  ax1 = plt.axes(xlim=(-20, 400), ylim=(0, 400))
  line, = ax1.plot([], [], lw=2)
    
  f = open(str(fileName[1]))
  i = 0
  nodes = [] 
  for line in f:
    nodes.append(line)
    i += 1
  print("totalNodes: "+str(len(nodes)))
  
  nodeContainer = [] #contains all the nodes
  for index in range(i):
    #initialize all the nodes to zero
    node = np.zeros((len(nodes[index].split())/3,3))
    nodeContainer.append(node)

    
  indexOfNodes = 0
  listOfCoords = []

  for element in nodes:
    indexOfCoords = 0
    #list of the coordenates of node "element" as [t0,x0,y0,t1...]
    listOfCoords = element.split()
    i = 0 #line of an specific node, so: [tn,xn,yn]
    while indexOfCoords != len(listOfCoords):
      nodeContainer[indexOfNodes][i][0] = listOfCoords[indexOfCoords]
      nodeContainer[indexOfNodes][i][1] = listOfCoords[indexOfCoords + 1]
      nodeContainer[indexOfNodes][i][2] = listOfCoords[indexOfCoords + 2]
      indexOfCoords += 3
      i += 1
    indexOfNodes += 1
  ax1 = fig1.add_subplot(1,1,1)
  legendList = []
  for i in range(len(nodeContainer)):
    ax1.plot(nodeContainer[i][:,1],nodeContainer[i][:,2], '--')
    legendList.append(str(i))
  for j in range(len(nodeContainer[0][:,1])):
    #annotates time in each data point for the first node
    if(j==0 or j==len(nodeContainer[0][:,1])-1 or j%50==0):
        ax1.annotate(str(nodeContainer[0][j][0]), (nodeContainer[0][j][1],nodeContainer[0][j][2]))  
  legend(legendList)
  #plt.show()
```
