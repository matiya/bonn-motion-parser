As can be seen, the dog's movement was simulated through a sine wave. The x axis in this sine function is the firefighter who moves in a straight line. Now, about the speed of the dogs: of course it should be randomized, but besides that, to make it so that the dogs follow the firefighters (i.e. they move through the y axis more or less together) it is necessary to calculate the length of the sine wave. This is done with the following integral:
$$DELTA\_X ~ *~ \int_a^{b} \sqrt{1+\cos^2 x}\,\mathrm dx$$ I don't know how to do this in c++, so unfortunately we'll have to go with some SAC and compute the result, like:

Introduced deviations from the sine wave. Every time a dog updates it's position there is a probability of it straying away from the path. Looks a little bit rough because there are no intermidiate points between the deviation and the dog's current position as can be seen in the fact that not all the dogs end up in the same y position despite having the same speed.

To control the dogs' speed and position when gone astray: say the dog's current position is (x,y) and the goal position is (xF, yF), then a versor should be calculated, like this
$$ \begin{align}
xDir = xF - x\\
yDir = yF - y\\
xVersor = xDir / \sqrt{ xDir^2 + yDir^2} \\
yVersor = yDir / \sqrt{ xDir^2 + yDir^2}\\
\end{align}$$
Then, to reach the final position in each timeStep: $$ nextX = previousX + timeStep * nodeSpeed * xVersor$$ and the same goes for nextY

In order to make the speed of the dogs along the path behave nonuniformly, these are drawn from a cuasi normal distribution, with mean in DOG_SPEED and a standar deviation given by the parameter STD_DEVIATION.
It should be noted that since the probability of drawing a really high speed is not null in a normal distribution, all the numbers drawn lie between (DOG_SPEED - STD_DEVIATION, DOG_SPEED + STD_DEVIATION). Hence, cuasi normal distrubution.
