```python
import pandas as pd
import numpy as np
```

### Import x axis values for ff from data

```python
number_of_runs = 100
columns = len( open("/home/default/workspace/data/run0").readline().split() )
usable_cols = range(1,columns, 4)
skipable_rows = range(1,12,2)
indexes = range(0,len(usable_cols))
data_list = []
for i in range(number_of_runs):
    dataF = pd.read_table( open("/home/default/workspace/data/run"+str(i),'r'), 
                          delim_whitespace = True,
                          usecols = usable_cols, 
                          skiprows = skipable_rows,
                          names = indexes,
                          index_col = None
                          ) 
    for i in range(1,len(usable_cols),4):
        del dataF[i]
    data_list.append( dataF )
```

### Get std deviation for a serie of x axis values

#### In the table below, the columns are the nodes and the rows are the simulation number

```python
mean_list = []
mean_df = pd.DataFrame()
for element in data_list:
    mean_list.append( element.transpose().describe().loc["std"] )
df = pd.DataFrame(mean_list)
df.set_index([range(number_of_runs)])
```
