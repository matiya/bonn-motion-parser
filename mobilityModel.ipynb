### Position

The following parameters determine the initial conditions: <br>
<h6>initial{x,y}</h6>
<em>default:</em>
<code>
INITIAL_X = 0
INITIAL_Y = 0
</code>
Set the initial x and y position <strong>for the first node</strong>. I.e. the one furthest to the left. Of course, the rest will be arranged according to the delta{X,Y} parameters if set or evenly spaced across the horizontal axis if not, like:

<h6>final{x,y}</h6>
<em>default:</em>
<code>
finalPosition.x = 0
finalPosition.y = 0
</code>
The direction the nodes will travel to. An angle is derived from these two parameters as:
<code>
ANGLE = std::atan2(finalPosition.y, finalPosition.x)
</code>
More info [here](#Rotation).<br>
<strong>[W]</strong> Do note that once this position is reached the simulation will not end. The execution  time is only limited by the <strong>duration</strong> parameter.

<h6>delta{X,Y}</h6>
<em>default:</em>
<code>
DELTA_X = FIELD_SIZE_X / NUM_NODES
DELTA_Y = 0
</code>
How spaced the nodes are in between them. The position is relative to the leftmost node, so if deltaX = 20 and deltaY = -10 and the node with index 0 is in (0,0) the node with index 1 will be on (20, -10) the node with index 2 will be in (40, -20) and the node with index n will be in (n*deltaX, n*deltaY).

<h6>amplitude</h6>
<em>default:</em>
<code>
AMPLITUDE = FIELD_SIZE_X / (NUM_NODES * 2);
if(AMPLITUDE > 50)
  AMPLITUDE = 50;
</code>
Amplitude of the sine wave. <br>
It is also considered when adjusting the speed of a dog if it has gone too far from firefighter. It is considered ok for a dog to be at a distance (firefighterPosition + AMPLITUDE/2, firefighterPosition + AMPLITUDE), in other words it will be always ahead of firefighter by at least AMPLITUDE/2 but not more than AMPLITUDE.

### Sine movement

As can be seen, the dog's movement was simulated through a sine wave, except it's not a proper sine wave but rather a set of discrete points aiming at a wrapper function wich is an actual sine wave. <br>
It can be seen below that there is a parametrized sine wave which parameter is timeStep. This wave acts as a wrapper for the actual movement of the dogs. The dogs then follow (or try to follow ) the sine wave, calculating a unit vector from their current position to the wrapper function.

$$ \begin{align}
xDir = xWrapper - xCurrent\\
yDir = yWrapper - yCurrent\\
xVersor = xDir / \sqrt{ xDir^2 + yDir^2} \\
yVersor = yDir / \sqrt{ xDir^2 + yDir^2}\\
\end{align}$$
Then, to reach the final position in each timeStep: 
$$ nextX = previousX + timeStep * nodeSpeed * xVersor \\
nextY = previousY + timeStep * nodeSpeed * yVersor $$

While it might sound overly complicated, it's actually simpler to use a wrapper function when there is a need to operate on the movement, for instance, when rotating or adding, substaction, acceleration, etc.
As the nodes move through timesteps, it goes without saying that the shorter they are, the more the nodes' movement will resemble a proper sine wave.


<a id='Deviations'> </a>

### Deviations

Introduced deviations from the sine wave. Every time a dog updates it's position there is a probability of it straying away from the path a random destination is selected as long as it lies outside the line of sight of the node (otherwise it might swerve 3 meters) and within the boundary set in the parameters.

To control the dogs' speed and position when gone astray a similar procedure to the sine wave is done: Get a unit vector to the random destination and move each timeStep towards it and then return to the position of the firefighter

### Obstacles

An obstacle is defined as a line segment given by two points (x1,y1) and (x2,y2). Several line segments can form a polygon or not, that's left to the user. 
An obstacle definition is given in a separated file (-o flag) as a list of floats delimited by commas. Like this:


That would define two line segments in (1.0;2.3), (3.1,4) and (5,6);(7,8).\n
Any node would have to notice an obstacle as soon as it enters its line of sight (set by parameter LOS), then ask itself if given its current movement, a collision might happen. If yes, calculate the shortest path to avoid collision. <br>
While avoiding an obstacle, both ff and dogs leave their charachteristic movement and advance together at the same speed in a linear way to the nearest edge of the obstacle. 

If there are no obstacles in the new calculated position, the ff will head in that direction. It will then try to return to its original position and if not possible (because there is an obstacle in the way) it will continue advancing straight to the goal position until the way is clear. 

For the time being and until I decide in a better, more meaningful way to check field size limits, these will not be considered when deciding the path. It is then perfectly possible for a node to go beyond FIELD_SIZE_X or FIELD_SIZE_Y. At least for now.

## Speed 

To make it so that the dogs follow the firefighters (i.e. they move through the y axis more or less together) it is necessary to calculate the length of the sine wave. This is done with the following integral:
$$DELTA\_X ~ *~ \int_a^{b} \sqrt{1+\cos^2 x}\,\mathrm dx$$ 

I don't know how to do this in c++, so unfortunately we'll have to go with <a href ="http://www.wolframalpha.com/input/?i=5*%28int+sqrt%281%2B%28cos%28x%29%29^2%29+from+x%3D0+to+x%3D2pi%29"> some SAC </a> and compute the result.



The speed is then derived from a normal distribution with a mean and a standard deviation given by the user, as:<br>
$$ getSpeed(mean, standar\_deviation)$$

It should be noted that since the probability of drawing a really high speed is not null in a normal distribution, all the numbers drawn lie between $(DOG\_SPEED - STD\_DEVIATION, DOG\_SPEED + STD\_DEVIATION)$. Hence cuasi normal distribution.

PROBLEM: If the speed's variation between two sucessive time points is too sudden, it can happen that:
$$y(t) = time * speed1 = 100s * 3.1m/s  = 310 m \\
y(t+1) = (time + dt) * speed2  = 101s * 2m/s  = 202 m $$

And the node would suddendly lie behind. Therefore the change in speed at least:

$$ speed2 >= \frac{time * speed1}{time + dt} $$ 

Of course, as time >> t  the ecuation above becomes speed2 == speed1, as can be seen in the graph below.

## Rotation

Rotation is achieved in 3 steps: <br>
1) Parametrizating the sine curve as:
$$ px = x(t=0) + DELTA_X * sin(py/8) \\
    py =  t * vel(t) $$
where t is time, DELTA_X is the amplitude of the sine wave and vel(t) is a function that describes the velocity. <br>
2) Multiplying by a rotation matrix:

$$   \begin{pmatrix}
  rx \\
  ry \\
 \end{pmatrix} =
\begin{pmatrix}
  cos(angle) & -sin(angle)\\
  sin(angle) & cos(angle)\\
 \end{pmatrix}  \times
  \begin{pmatrix}
  px \\
  py \\
 \end{pmatrix} $$

$$ rx = px * cos(angle) - py * sin(angle) \\
ry = px * sin(angle) + py * cos(angle) $$

3) Translating from the origin:
 
 $$ tx = px * cos(angle) - py * sin(angle) + x(0)\\
ty = px * sin(angle) + py * cos(angle)  + y(0)$$


That would the ideal scenario, the problem is that there is no $vel(t)$ to derive the position in $y$ as function of time. This is because the speed of the nodes is random around a mean. So if i say $py = t $ it works, as seen below, but the speed is not taken into account. That is unacceptable and would much diminish the quality of the simulation, so I'm going to explore a method of obtaining said function. <br>
I don't think RK4 or verlet are possible, because as stated before, the velocity is random and thus it can't be expressed as a DE.

```python
import sys
import os
import getopt
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.lines import Line2D
from matplotlib import animation
from IPython.display import Image

def oneNode(opt, fileName):
  """plot one node"""
  ax = fig.add_subplot(1,1,1)
  a = np.loadtxt(str(fileName[1]))
  x = a[:,0]
  y = a[:,1]
  ax.plot(x,y,'--')
  if opt == 1:
    plt.show()
    
def allNodesIn(cwd):
  """plots all the .visplotX files found in the given dir
  """
  legendList = []
  print("found the following:")
  for argument in os.listdir(cwd):
    if 'visplot' in argument:
      oneNode(2,[1,str(cwd)+"/"+argument])
      print(str(cwd)+"/"+argument)
      legendList.append(str(argument))
  ax = fig.add_subplot(1,1,1)
  ax.legend(legendList)
  plt.draw()
  plt.show()

def allNodes(fileName, opt = ""):
  """plots all the nodes, the input is the decompressed file produced by bm"""
  fig1 = plt.figure(figsize=(16,8))
  #open trace file and obstacle file  
  f = open(str(fileName[0]),'r')
  obsf = open(str(fileName[1]),'r')
    
  i = 0
  j = 0
  k = 0
  #empty containers for I don't even know anymore  
  nodes = [] 
  dataObs = []
  dataObs1 = []
    
  for line in f:
    nodes.append(line)
    i += 1
  print("totalNodes: "+str(len(nodes)))
  #parse and plot obstacles
  for line in obsf:
    j=0
    for element in line.strip().split(','):
      if(j%2==0):
        try:
          dataObs.append( float(element) )
        except:
          pass
      else:
        try:
          dataObs1.append( float(element) )
        except:
          pass
      j+=1
  #plot obstacles
  for k in range(len(dataObs)):
    if k%2==0:
      plt.plot(dataObs[k:k+2],dataObs1[k:k+2], 'k--', lw = 3)
  
  
  nodeContainer = [] #contains all the nodes
  for index in range(i):
    #initialize all the nodes to zero
    node = np.zeros((len(nodes[index].split())/4,4))
    nodeContainer.append(node)

    
  indexOfNodes = 0
  listOfCoords = []

  for element in nodes:
    indexOfCoords = 0
    #list of the coordenates of node "element" as [t0,x0,y0,t1...]
    listOfCoords = element.split()
    i = 0 #line of an specific node, so: [tn,xn,yn]
    while indexOfCoords != len(listOfCoords):
      nodeContainer[indexOfNodes][i][0] = listOfCoords[indexOfCoords]
      nodeContainer[indexOfNodes][i][1] = listOfCoords[indexOfCoords + 1]
      nodeContainer[indexOfNodes][i][2] = listOfCoords[indexOfCoords + 2]
      nodeContainer[indexOfNodes][i][3] = listOfCoords[indexOfCoords + 3]
      indexOfCoords += 4
      i += 1
    indexOfNodes += 1
  ax1 = fig1.add_subplot(1,1,1)
  ax1.set_ylabel("y position")
  ax1.set_xlabel("x position")
  legendList = []
  for i in range(len(nodeContainer)):
    ax1.plot(nodeContainer[i][:,1],nodeContainer[i][:,2], '-')
    legendList.append(str(i))
  if(opt == "time"):  
      for j in range(len(nodeContainer[0][:,1])):
        #annotates time in each data point for the first node
        if(j==0 or j==len(nodeContainer[0][:,1])-1 or j%50==0):
            ax1.annotate(str(nodeContainer[0][j][0]), (nodeContainer[0][j][1],nodeContainer[0][j][2]))  
  legend(legendList)
  #plt.show()
```

```python
"""load a trace file from disk and plot a selected column
against number of timeSteps, where mode can be either
speed, x or y position.
"""
def plotVar(filename, mode) :
    columns = len( open(filename).readline().split() )
    if(mode == "speed"):
        usable_cols = range(3,columns, 4)
    elif(mode == "x"): 
        usable_cols = range(1,columns, 4)
    elif(mode == "y"): 
        usable_cols = range(2,columns, 4)
    skipable_rows = range(0,12,2)
    indexes = range(0,len(usable_cols))
    data_list = []
    dataF = pd.read_table( open(filename,'r'), 
                        delim_whitespace = True,
                        usecols = usable_cols, 
                        skiprows = skipable_rows,
                        names = indexes,
                        index_col = None
                        ) 
    if(mode == "speed"):
        for i in range(3,len(usable_cols),4):
            del dataF[i]
        del dataF[0] #delete initial speed, usually an off value
    elif(mode == "x"): 
        for i in range(1,len(usable_cols),4):
            del dataF[i]
    elif(mode == "y"): 
        for i in range(2,len(usable_cols),4):
            del dataF[i]
    minValue = dataF.min().min()
    maxValue = dataF.max().max()
    figure(figsize=(16,10))    
    a = dataF.transpose()
    ax1 = plt.subplot(2,2,1, xlim= (0,20), ylim=(minValue, maxValue))
    ax1.set_ylabel(mode)
    ax1.set_xlabel("samples")
    plt.plot(a, axes=ax1)
    ax2 = plt.subplot(2,2,2, xlim= (0,100), ylim=(minValue, maxValue))
    ax2.set_ylabel(mode)
    ax2.set_xlabel("samples")
    plt.plot(a, axes=ax2)
    ax3 = plt.subplot(2,2,3, xlim= (300,350), ylim=(minValue, maxValue))
    ax3.set_ylabel(mode)
    ax3.set_xlabel("samples")
    plt.plot(a, axes=ax3)
    ax4 = plt.subplot(2,2,4, xlim= (0,700), ylim=(minValue, maxValue))
    ax4.set_xlabel("samples")
    plt.plot(a, axes=ax4)
    plt.show()
    return dataF
```
